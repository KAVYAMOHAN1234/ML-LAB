```python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
n = 5
for i in range(n):
    print(" " * (n - i - 1), end="")
    print("* " * (i + 1))
```
